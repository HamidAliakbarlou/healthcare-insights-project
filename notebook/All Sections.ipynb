#### **Treating Missing Values (less than 5%)**

For some columns that the percentage of null values are less than 5%, we can either drop the records or impute, and for those have more than 5%, we should impute using methods like "Mode", "Mean", "Random Forest".

For now, Let’s impute the null values with the Mode (Most Frequent Category) for columns: "age", "gender", "cost", "bmi", "cholesterol_total"

```python
# Find the most frequent category (mode)
mode_value = df['age'].mode()[0]
mode_value2 = df['gender'].mode()[0]

# Fill missing values with the mode
df['age'] = df['age'].fillna(mode_value)
df['gender'] = df['gender'].fillna(mode_value2)
```

```python
# Find the most frequent category (mode)
mode_value = df2['cost'].mode()[0]

# Fill missing values with the mode
df2['cost'] = df2['cost'].fillna(mode_value)
```

```python
# Find the most frequent category (mode)
mode_value = df3['bmi'].mode()[0]
mode_value2 = df3['cholesterol_total'].mode()[0]

# Fill missing values with the mode
df3['bmi'] = df3['bmi'].fillna(mode_value)
df3['cholesterol_total'] = df3['cholesterol_total'].fillna(mode_value2)
```

#3. Risk Factors Insights and Statistics

## 3.1. Risk Factors for Diabetes

Find any risk factors (associations): Are there any strong correlations between patienst with diabetes vs BMI, blood pressure, and cholesterol?

**Step 1**: Merge the 3 Tables

```python
# Merge patients + medical records
df_merged1 = pd.merge(df, df2, on='patient_id', how='inner')

# Merge with biometric data
merged_df = pd.merge(df_merged1, df3, on='patient_id', how='inner')
```

```python
merged_df.shape
```

**Step 2**: Create Binary Target for Diabetes

```python
# Create a binary column:
merged_df['has_diabetes'] = merged_df['diagnosis'].str.contains('diabetes', case=False, na=False).astype(int)
```

**Step 3**: Create Visual Boxplots for Numeric Features

```python
features = ['bmi', 'cholesterol_total', 'blood_pressure_systolic', 'blood_pressure_diastolic', 'age']

# Plot boxplots for each feature vs diabetes
for feature in features:
    plt.figure(figsize=(6, 4))
    #sns.boxplot(data=merged_df, x='has_diabetes', y=feature, palette='pastel', width=0.5,)
    sns.boxplot(data=merged_df, x='has_diabetes', y=feature, width=0.6)

    plt.title(f'{feature} by Diabetes Status', fontsize=15)
    plt.xlabel('Diabetes (0=No, 1=Yes)', fontsize=13)
    plt.ylabel(feature, fontsize=13)

    plt.xticks(fontsize=11)
    plt.yticks(fontsize=11)
    plt.tight_layout()
    plt.grid(True, linestyle='--', alpha=0.3)
    sns.despine()
    plt.show()
```



**Step 4**: T-Test to See if Groups Are Statistically Different

-  **What this tells us:** It indicates If the average value for that feature is significantly different between diabetics and non-diabetics.



**Assumption:**

If the BMI values for both groups (diabetic and non-diabetic) are approximately normally distributed and their variances are similar, then the appropriate test is the Independent Samples t-test (also called the two-sample t-test).



```python
from scipy.stats import ttest_ind

for feature in features:
    group1 = merged_df[merged_df['has_diabetes'] == 1][feature]
    group0 = merged_df[merged_df['has_diabetes'] == 0][feature]
    t_stat, p_val = ttest_ind(group1, group0, equal_var=False, nan_policy='omit')
    print(f"{feature}: p-value = {p_val:.4f}")
```

**Check the above-mentioned Assumption of the t-test:**

-  Both groups are independent

-  BMI in each group is normally distributed (use shapiro() or sns.histplot() to check)

```python
from scipy.stats import shapiro

# Subsets
bmi_diabetic = merged_df[merged_df['has_diabetes'] == 1]['bmi']
bmi_nondiabetic = merged_df[merged_df['has_diabetes'] == 0]['bmi']

# Normality test
shapiro(bmi_diabetic)
shapiro(bmi_nondiabetic)
```

**Result**

- BMI and age show statistically significant associations with diabetes at the 95% confidence level (p < 0.05), confirming what we observed in the boxplots.

- Cholesterol and diastolic blood pressure do not show significant effects (p > 0.3), indicating weak or no relationship with diabetes based on both visuals and statistical tests.

###  Check whether individuals with a higher BMI are more likely to have diabetes (One sided Hypothesis test)


We can do it manually by halving the p-value (if the direction of the difference matches our expectation):

```python
feature = 'bmi'
group1 = merged_df[merged_df['has_diabetes'] == 1][feature]
group0 = merged_df[merged_df['has_diabetes'] == 0][feature]

t_stat, p_val_two_sided = ttest_ind(group1, group0, equal_var=False, nan_policy='omit')

# Check if group1 mean is greater than group0
if group1.mean() > group0.mean():
    p_val_one_sided = p_val_two_sided / 2
else:
    p_val_one_sided = 1 - (p_val_two_sided / 2)

print(f"One-sided p-value: {p_val_one_sided:.4f}")
```
